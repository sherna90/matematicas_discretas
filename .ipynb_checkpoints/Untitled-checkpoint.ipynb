```python
import numpy as np

perm = lambda n,r : np.math.factorial(n)/np.math.factorial(n-r)
```

```python
print ((26-8)**4)*(10**2)

print perm(4,1)
```

```python
def simple_generator_function(n):
    num=0
    if num<n:
        num=num+1
    yield num

for value in simple_generator_function(10):
    print(value)
```
